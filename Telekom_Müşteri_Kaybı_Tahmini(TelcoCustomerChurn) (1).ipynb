## 5.Aykırı Değer Kontrolü

```python
# Aykırı değerleri tespit etmek için IQR yöntemini kullanacağız.

def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.25)
    quartile3 = dataframe[variable].quantile(0.75)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False

# Sayısal değişkenler için aykırı değer kontrolü
numeric_cols = df.select_dtypes(include=['float', 'int']).columns

for col in numeric_cols:
    print(col, check_outlier(df, col))

# Aykırı değerleri baskılamak için bir fonksiyon tanımlayalım
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit

# Aykırı değerleri baskılayalım
for col in numeric_cols:
    if check_outlier(df, col):
        replace_with_thresholds(df, col)
        print(f"{col} sütunundaki aykırı değerler baskılandı.")
```

## 7.Veri Modellemeye Giriş

**Kategorik değişkenleri sayısal değerlere dönüştürelim ve veriyi modellemeye hazır hale getirelim.**

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Get a list of all categorical features in the DataFrame
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Apply Label Encoding to all categorical features
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

# Create a ColumnTransformer to apply OneHotEncoder to all categorical features
categorical_features = categorical_cols.tolist()
numerical_features = df.select_dtypes(exclude=['object', 'category']).columns.tolist()
numerical_features.remove('Churn')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
    ])

# Apply the preprocessing to the data
X = preprocessor.fit_transform(df.drop('Churn', axis=1))
y = df['Churn']

# Veriyi eğitim ve test setlerine ayırma
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Sınıflandırma modelleri geliştirip performanslarını değerlendirelim.**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

# XGBoost
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(confusion_matrix(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
# Model performanslarını karşılaştırma
models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
accuracies = [accuracy_score(y_test, y_pred_log), accuracy_score(y_test, y_pred_dt),
              accuracy_score(y_test, y_pred_rf), accuracy_score(y_test, y_pred_xgb)]

sns.barplot(x=models, y=accuracies)
plt.ylim(0.7, 1.0)
plt.ylabel('Accuracy')
plt.title('Model Performans Karşılaştırması')
plt.show()

```

## 8.Sonuçlar

Kampanya ve Teklifler için Öneriler

Yüksek riskli müşterilere yönelik özel kampanyalar veya teklifler sunabiliriz.

Örneğin:

*İndirimler*: Yüksek riskli müşterilere özel indirimler sunulabilir.

*Kişiselleştirilmiş Teklifler*: Müşterilerin kullanım alışkanlıklarına göre özel teklifler (örneğin, daha fazla internet kotası).

*Müşteri Destek Hizmetleri*: Bu müşterilere özel müşteri destek hizmetleri sunulabilir.

*Sadakat Programları*: Sadakat programlarına dahil edilerek, müşterilerin bağlılığı artırılabilir.

```python
# Random Forest modelinden churn olasılıklarını alma
rf_probs = rf.predict_proba(X_test)[:, 1]  # Churn olasılıkları (1 sınıfı için)
rf_churn_risk = pd.DataFrame({
    'ChurnProbability': rf_probs
})

# Churn olasılığı yüksek olan müşterileri sıralama
high_risk_customers = rf_churn_risk[rf_churn_risk['ChurnProbability'] > 0.7].sort_values(by='ChurnProbability', ascending=False)
print("Churn Risk High Customers:")
print(high_risk_customers)
```

```python
# Müşterileri churn riskine göre segmentlere ayırma
rf_churn_risk['RiskLevel'] = pd.cut(
    rf_churn_risk['ChurnProbability'],
    bins=[0, 0.3, 0.7, 1],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

print(rf_churn_risk.groupby('RiskLevel').size())
```
